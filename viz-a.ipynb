# Data Visulization For Cars

```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('mpg')
```

```python
df.head()
```

```python
sns.relplot(x='horsepower', y='acceleration', hue=df.origin.tolist(), data=df);
```

```python

year_mean = df.groupby('cylinders', as_index=False).mean()

year_mean
```

```python
graph = sns.barplot(x="cylinders", y="mpg", data=year_mean, ci=None)
```

```python
sns.lmplot(x="cylinders", y="mpg", data=year_mean);
```

```python

graph = sns.relplot(x="horsepower", y="acceleration", hue=df.model_year.tolist(), data=df)
```
